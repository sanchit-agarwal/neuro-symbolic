```python
# Program to plot training times for all the models using log files

from os import listdir
from os.path import isfile, join
import pandas as pd
import matplotlib.pyplot as plt

path = "./"
log_files = []

for file in listdir(path):
    if isfile(join(path, file)) and file.endswith(".log"):
        log_files.append(file)       
        
accuracies = []        
plt.figure()
            
for file in log_files:
    print(file)
    df = pd.read_csv(file)
    if "cnn" in file:
        label = file[:3].upper()
    else:
        label = file[:2]
        
    plt.plot(df["time"], df["loss"], label=label)
    
    plt.legend(loc="best")

plt.xlabel("total time(secs)")
plt.ylabel("loss")
plt.savefig("images/training.png")
```
